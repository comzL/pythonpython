```python
# 首先安装必要的库（如果尚未安装）
!pip install requests pandas matplotlib

import requests
import json
import time
import random
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

class NetEaseMusicCrawler:
    def __init__(self):
        self.headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36',
            'Referer': 'https://music.163.com/',
            'Origin': 'https://music.163.com'
        }
        self.base_url = 'https://music.163.com/api/v1/resource/comments/R_SO_4_{}'
        
    def get_comments(self, song_id, limit=100, offset=0):
        """获取指定歌曲的评论"""
        url = self.base_url.format(song_id)
        params = {'limit': limit, 'offset': offset}
        
        try:
            response = requests.get(url, headers=self.headers, params=params)
            if response.status_code == 200:
                return response.json().get('comments', [])
            print(f"⚠️ 请求失败，状态码: {response.status_code}")
            return []
        except Exception as e:
            print(f"❌ 发生错误: {e}")
            return []
    
    def get_all_comments(self, song_id, max_comments=500):
        """获取指定歌曲的所有评论"""
        all_comments = []
        limit = 100
        offset = 0
        
        with requests.Session() as session:
            while len(all_comments) < max_comments:
                comments = self.get_comments(song_id, limit, offset)
                if not comments:
                    break
                    
                all_comments.extend(comments)
                offset += len(comments)
                
                display(Markdown(f"🔄 已获取 {len(all_comments)}/{max_comments} 条评论..."))
                
                # Jupyter中可以使用更短的延迟
                time.sleep(random.uniform(0.3, 0.8))
        
        return all_comments[:max_comments]
    
    def comments_to_dataframe(self, comments):
        """将评论列表转换为DataFrame"""
        data = []
        for comment in comments:
            data.append({
                '用户': comment['user']['nickname'],
                '用户ID': comment['user']['userId'],
                '评论内容': comment['content'],
                '点赞数': comment['likedCount'],
                '评论时间': pd.to_datetime(comment['time'], unit='ms'),
                'IP属地': comment.get('ipLocation', {}).get('location', '未知')
            })
        return pd.DataFrame(data)

# 在Jupyter中创建实例并运行
crawler = NetEaseMusicCrawler()

# 设置歌曲ID和最大评论数
song_id = 165339  # 汪苏泷《万有引力》
max_comments = 200

# 获取评论
display(Markdown(f"## 🎵 开始爬取歌曲ID {song_id} 的评论..."))
comments = crawler.get_all_comments(song_id, max_comments)

if comments:
    # 转换为DataFrame
    df = crawler.comments_to_dataframe(comments)
    
    # 显示前5条评论
    display(Markdown("### 📋 前5条评论:"))
    display(df.head())
    
    # 简单分析
    display(Markdown("### 📊 基本统计:"))
    display(df.describe())
    
    # 点赞数分布可视化
    plt.figure(figsize=(10, 5))
    df['点赞数'].hist(bins=20)
    plt.title('评论点赞数分布')
    plt.xlabel('点赞数')
    plt.ylabel('评论数量')
    plt.show()
    
    # 保存为CSV
    csv_file = f'netease_comments_{song_id}.csv'
    df.to_csv(csv_file, index=False, encoding='utf-8-sig')
    display(Markdown(f"💾 评论已保存为 {csv_file}"))
    
    # 显示一些有趣的评论
    top_liked = df.sort_values('点赞数', ascending=False).head(3)
    display(Markdown("### 🏆 点赞最高的3条评论:"))
    for idx, row in top_liked.iterrows():
        display(Markdown(f"**{row['用户']}** (👍 {row['点赞数']}):\n> {row['评论内容']}\n"))
else:
    display(Markdown("❌ 未能获取任何评论"))
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import jieba
import jieba.analyse
from gensim import corpora, models
import pyLDAvis
import pyLDAvis.gensim_models
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import Counter
import zhconv  # 简繁转换

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体显示中文
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 创建结果保存目录
result_dir = "D:/result"
os.makedirs(result_dir, exist_ok=True)

# 1. 数据加载与预处理
def load_and_preprocess_data(filepath):
    df = pd.read_csv(filepath)
    
    # 数据清洗
    df['评论内容'] = df['评论内容'].str.replace(r'[^\w\s]', '', regex=True)  # 去除标点
    df['评论内容'] = df['评论内容'].str.replace('\n', ' ')  # 去除换行符
    df['评论内容'] = df['评论内容'].apply(lambda x: zhconv.convert(x, "zh-cn"))    
    # 时间处理
    df['评论时间'] = pd.to_datetime(df['评论时间'])
    df['评论年份'] = df['评论时间'].dt.year
    df['评论月份'] = df['评论时间'].dt.to_period('M')
    
    return df

# 2. 中文文本处理
def setup_chinese_analyzer():
    # 加载停用词
    stopwords = set()
    with open('chinese_stopwords.txt', 'r', encoding='utf-8') as f:
        for line in f:
            stopwords.add(line.strip())
    
    # 添加音乐领域词典
    music_terms = ['副歌', '编曲', '旋律', '节奏', '和弦', '前奏', '间奏', '尾奏', 
                  '音色', '歌词', '万有引力', '汪苏泷', '主歌', 'bridge', 'hook']
    for term in music_terms:
        jieba.add_word(term)
    
    return stopwords

def chinese_text_processing(text, stopwords):
    words = jieba.cut(text)
    return [word for word in words if word not in stopwords and len(word) > 1]

# 3. TF-IDF关键词分析
def tfidf_analysis(df, stopwords, save_dir):
    # 准备文本数据
    df['分词内容'] = df['评论内容'].apply(lambda x: chinese_text_processing(x, stopwords))
    df['分词文本'] = df['分词内容'].apply(' '.join)
    
    # TF-IDF分析
    tfidf = TfidfVectorizer(max_features=100)
    tfidf_matrix = tfidf.fit_transform(df['分词文本'])
    feature_names = tfidf.get_feature_names_out()
    
    # 获取关键词权重
    word_scores = {word: tfidf_matrix[:, idx].sum() for idx, word in enumerate(feature_names)}
    top_words = sorted(word_scores.items(), key=lambda x: x[1], reverse=True)[:20]
    
    # 保存关键词结果
    keywords_df = pd.DataFrame(top_words, columns=['关键词', 'TF-IDF权重'])
    keywords_df.to_csv(f'{save_dir}/tfidf_keywords.csv', index=False, encoding='utf-8-sig')
    
    # 可视化
    plt.figure(figsize=(12, 8))
    sns.barplot(x='TF-IDF权重', y='关键词', data=keywords_df)
    plt.title('TF-IDF权重最高的20个关键词')
    plt.tight_layout()
    plt.savefig(f'{save_dir}/tfidf_top_keywords.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    # 生成词云
    all_words = ' '.join(df['分词文本'])
    wordcloud = WordCloud(width=800, height=400, 
                         background_color='white',
                         font_path='simhei.ttf',
                         max_words=100).generate(all_words)
    
    plt.figure(figsize=(15, 8))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('评论关键词词云')
    plt.savefig(f'{save_dir}/word_cloud.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    return df, feature_names, tfidf_matrix

# 4. LDA主题建模
def lda_topic_modeling(df, save_dir, num_topics=4):
    # 准备LDA输入数据
    texts = df['分词内容'].tolist()
    dictionary = corpora.Dictionary(texts)
    corpus = [dictionary.doc2bow(text) for text in texts]
    
    # 训练LDA模型
    lda_model = models.LdaModel(corpus=corpus,
                               id2word=dictionary,
                               num_topics=num_topics,
                               random_state=42,
                               passes=15,
                               alpha='auto')
    
    # 保存主题关键词
    topics = lda_model.print_topics(num_words=8)
    with open(f'{save_dir}/lda_topics.txt', 'w', encoding='utf-8') as f:
        for idx, topic in topics:
            f.write(f"主题{idx}: {topic}\n")
    
    # 可视化主题
    vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary)
    pyLDAvis.save_html(vis, f'{save_dir}/lda_visualization.html')
    
    # 为每条评论分配主要主题
    df['主要主题'] = [sorted(lda_model[doc], key=lambda x: x[1], reverse=True)[0][0] for doc in corpus]
    
    # 主题分布可视化
    topic_dist = df['主要主题'].value_counts().sort_index()
    plt.figure(figsize=(10, 6))
    topic_dist.plot(kind='bar')
    plt.title('评论主题分布')
    plt.xlabel('主题编号')
    plt.ylabel('评论数量')
    plt.savefig(f'{save_dir}/topic_distribution.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    return df, lda_model

# 5. 情感分析 (模拟)
def sentiment_analysis(df, save_dir):
    # 模拟情感分析 (实际应用中可用BosonNLP或其他API)
    np.random.seed(42)
    df['情感分数'] = np.random.normal(loc=0.6, scale=0.2, size=len(df))
    df['情感分数'] = df['情感分数'].clip(0, 1)  # 限制在0-1范围
    
    # 情感分布可视化
    plt.figure(figsize=(10, 6))
    sns.histplot(df['情感分数'], bins=20, kde=True)
    plt.title('评论情感分数分布')
    plt.xlabel('情感分数 (0-1)')
    plt.ylabel('频数')
    plt.axvline(0.5, color='r', linestyle='--')
    plt.text(0.52, plt.ylim()[1]*0.9, '中性阈值', color='r')
    plt.savefig(f'{save_dir}/sentiment_distribution.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    # 情感与主题的关系
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='主要主题', y='情感分数', data=df)
    plt.title('不同主题的情感分数分布')
    plt.xlabel('主题编号')
    plt.ylabel('情感分数')
    plt.savefig(f'{save_dir}/sentiment_by_topic.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    # 情感与点赞数的关系
    plt.figure(figsize=(12, 6))
    sns.scatterplot(x='情感分数', y='点赞数', data=df, alpha=0.6)
    plt.title('情感分数与点赞数的关系')
    plt.xlabel('情感分数')
    plt.ylabel('点赞数')
    plt.savefig(f'{save_dir}/sentiment_vs_likes.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    return df

# 6. 时间序列分析
def temporal_analysis(df, save_dir):
    # 按月份统计评论数量
    monthly_counts = df.groupby('评论月份').size()
    
    plt.figure(figsize=(14, 6))
    monthly_counts.plot()
    plt.title('评论数量随时间变化')
    plt.xlabel('时间')
    plt.ylabel('评论数量')
    plt.grid(True)
    plt.savefig(f'{save_dir}/comments_over_time.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    # 主题随时间变化
    monthly_topics = df.groupby(['评论月份', '主要主题']).size().unstack()
    monthly_topics.plot(kind='area', stacked=True, figsize=(14, 6))
    plt.title('主题随时间变化')
    plt.xlabel('时间')
    plt.ylabel('评论数量')
    plt.legend(title='主题')
    plt.savefig(f'{save_dir}/topics_over_time.png', dpi=300, bbox_inches='tight')
    plt.close()

# 主函数
def main():
    # 加载数据
    filepath = 'netease_comments_165339.csv'
    df = load_and_preprocess_data(filepath)
    
    # 设置中文分析器
    stopwords = setup_chinese_analyzer()
    
    # 执行分析流程
    df, feature_names, tfidf_matrix = tfidf_analysis(df, stopwords, result_dir)
    df, lda_model = lda_topic_modeling(df, result_dir)
    df = sentiment_analysis(df, result_dir)
    temporal_analysis(df, result_dir)
    
    # 保存处理后的数据
    df.to_csv(f'{result_dir}/processed_comments.csv', index=False, encoding='utf-8-sig')
    print(f"分析完成，结果已保存到: {result_dir}")

if __name__ == "__main__":
    main()
```
