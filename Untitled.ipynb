```python
# È¶ñÂÖàÂÆâË£ÖÂøÖË¶ÅÁöÑÂ∫ìÔºàÂ¶ÇÊûúÂ∞öÊú™ÂÆâË£ÖÔºâ
!pip install requests pandas matplotlib

import requests
import json
import time
import random
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

class NetEaseMusicCrawler:
    def __init__(self):
        self.headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36',
            'Referer': 'https://music.163.com/',
            'Origin': 'https://music.163.com'
        }
        self.base_url = 'https://music.163.com/api/v1/resource/comments/R_SO_4_{}'
        
    def get_comments(self, song_id, limit=100, offset=0):
        """Ëé∑ÂèñÊåáÂÆöÊ≠åÊõ≤ÁöÑËØÑËÆ∫"""
        url = self.base_url.format(song_id)
        params = {'limit': limit, 'offset': offset}
        
        try:
            response = requests.get(url, headers=self.headers, params=params)
            if response.status_code == 200:
                return response.json().get('comments', [])
            print(f"‚ö†Ô∏è ËØ∑Ê±ÇÂ§±Ë¥•ÔºåÁä∂ÊÄÅÁ†Å: {response.status_code}")
            return []
        except Exception as e:
            print(f"‚ùå ÂèëÁîüÈîôËØØ: {e}")
            return []
    
    def get_all_comments(self, song_id, max_comments=500):
        """Ëé∑ÂèñÊåáÂÆöÊ≠åÊõ≤ÁöÑÊâÄÊúâËØÑËÆ∫"""
        all_comments = []
        limit = 100
        offset = 0
        
        with requests.Session() as session:
            while len(all_comments) < max_comments:
                comments = self.get_comments(song_id, limit, offset)
                if not comments:
                    break
                    
                all_comments.extend(comments)
                offset += len(comments)
                
                display(Markdown(f"üîÑ Â∑≤Ëé∑Âèñ {len(all_comments)}/{max_comments} Êù°ËØÑËÆ∫..."))
                
                # Jupyter‰∏≠ÂèØ‰ª•‰ΩøÁî®Êõ¥Áü≠ÁöÑÂª∂Ëøü
                time.sleep(random.uniform(0.3, 0.8))
        
        return all_comments[:max_comments]
    
    def comments_to_dataframe(self, comments):
        """Â∞ÜËØÑËÆ∫ÂàóË°®ËΩ¨Êç¢‰∏∫DataFrame"""
        data = []
        for comment in comments:
            data.append({
                'Áî®Êà∑': comment['user']['nickname'],
                'Áî®Êà∑ID': comment['user']['userId'],
                'ËØÑËÆ∫ÂÜÖÂÆπ': comment['content'],
                'ÁÇπËµûÊï∞': comment['likedCount'],
                'ËØÑËÆ∫Êó∂Èó¥': pd.to_datetime(comment['time'], unit='ms'),
                'IPÂ±ûÂú∞': comment.get('ipLocation', {}).get('location', 'Êú™Áü•')
            })
        return pd.DataFrame(data)

# Âú®Jupyter‰∏≠ÂàõÂª∫ÂÆû‰æãÂπ∂ËøêË°å
crawler = NetEaseMusicCrawler()

# ËÆæÁΩÆÊ≠åÊõ≤IDÂíåÊúÄÂ§ßËØÑËÆ∫Êï∞
song_id = 165339  # Ê±™ËãèÊ≥∑„Ää‰∏áÊúâÂºïÂäõ„Äã
max_comments = 200

# Ëé∑ÂèñËØÑËÆ∫
display(Markdown(f"## üéµ ÂºÄÂßãÁà¨ÂèñÊ≠åÊõ≤ID {song_id} ÁöÑËØÑËÆ∫..."))
comments = crawler.get_all_comments(song_id, max_comments)

if comments:
    # ËΩ¨Êç¢‰∏∫DataFrame
    df = crawler.comments_to_dataframe(comments)
    
    # ÊòæÁ§∫Ââç5Êù°ËØÑËÆ∫
    display(Markdown("### üìã Ââç5Êù°ËØÑËÆ∫:"))
    display(df.head())
    
    # ÁÆÄÂçïÂàÜÊûê
    display(Markdown("### üìä Âü∫Êú¨ÁªüËÆ°:"))
    display(df.describe())
    
    # ÁÇπËµûÊï∞ÂàÜÂ∏ÉÂèØËßÜÂåñ
    plt.figure(figsize=(10, 5))
    df['ÁÇπËµûÊï∞'].hist(bins=20)
    plt.title('ËØÑËÆ∫ÁÇπËµûÊï∞ÂàÜÂ∏É')
    plt.xlabel('ÁÇπËµûÊï∞')
    plt.ylabel('ËØÑËÆ∫Êï∞Èáè')
    plt.show()
    
    # ‰øùÂ≠ò‰∏∫CSV
    csv_file = f'netease_comments_{song_id}.csv'
    df.to_csv(csv_file, index=False, encoding='utf-8-sig')
    display(Markdown(f"üíæ ËØÑËÆ∫Â∑≤‰øùÂ≠ò‰∏∫ {csv_file}"))
    
    # ÊòæÁ§∫‰∏Ä‰∫õÊúâË∂£ÁöÑËØÑËÆ∫
    top_liked = df.sort_values('ÁÇπËµûÊï∞', ascending=False).head(3)
    display(Markdown("### üèÜ ÁÇπËµûÊúÄÈ´òÁöÑ3Êù°ËØÑËÆ∫:"))
    for idx, row in top_liked.iterrows():
        display(Markdown(f"**{row['Áî®Êà∑']}** (üëç {row['ÁÇπËµûÊï∞']}):\n> {row['ËØÑËÆ∫ÂÜÖÂÆπ']}\n"))
else:
    display(Markdown("‚ùå Êú™ËÉΩËé∑Âèñ‰ªª‰ΩïËØÑËÆ∫"))
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import jieba
import jieba.analyse
from gensim import corpora, models
import pyLDAvis
import pyLDAvis.gensim_models
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import Counter
import zhconv  # ÁÆÄÁπÅËΩ¨Êç¢

# ËÆæÁΩÆ‰∏≠ÊñáÂ≠ó‰Ωì
plt.rcParams['font.sans-serif'] = ['SimHei']  # ‰ΩøÁî®Èªë‰ΩìÊòæÁ§∫‰∏≠Êñá
plt.rcParams['axes.unicode_minus'] = False  # Ëß£ÂÜ≥Ë¥üÂè∑ÊòæÁ§∫ÈóÆÈ¢ò

# ÂàõÂª∫ÁªìÊûú‰øùÂ≠òÁõÆÂΩï
result_dir = "D:/result"
os.makedirs(result_dir, exist_ok=True)

# 1. Êï∞ÊçÆÂä†ËΩΩ‰∏éÈ¢ÑÂ§ÑÁêÜ
def load_and_preprocess_data(filepath):
    df = pd.read_csv(filepath)
    
    # Êï∞ÊçÆÊ∏ÖÊ¥ó
    df['ËØÑËÆ∫ÂÜÖÂÆπ'] = df['ËØÑËÆ∫ÂÜÖÂÆπ'].str.replace(r'[^\w\s]', '', regex=True)  # ÂéªÈô§Ê†áÁÇπ
    df['ËØÑËÆ∫ÂÜÖÂÆπ'] = df['ËØÑËÆ∫ÂÜÖÂÆπ'].str.replace('\n', ' ')  # ÂéªÈô§Êç¢Ë°åÁ¨¶
    df['ËØÑËÆ∫ÂÜÖÂÆπ'] = df['ËØÑËÆ∫ÂÜÖÂÆπ'].apply(lambda x: zhconv.convert(x, "zh-cn"))    
    # Êó∂Èó¥Â§ÑÁêÜ
    df['ËØÑËÆ∫Êó∂Èó¥'] = pd.to_datetime(df['ËØÑËÆ∫Êó∂Èó¥'])
    df['ËØÑËÆ∫Âπ¥‰ªΩ'] = df['ËØÑËÆ∫Êó∂Èó¥'].dt.year
    df['ËØÑËÆ∫Êúà‰ªΩ'] = df['ËØÑËÆ∫Êó∂Èó¥'].dt.to_period('M')
    
    return df

# 2. ‰∏≠ÊñáÊñáÊú¨Â§ÑÁêÜ
def setup_chinese_analyzer():
    # Âä†ËΩΩÂÅúÁî®ËØç
    stopwords = set()
    with open('chinese_stopwords.txt', 'r', encoding='utf-8') as f:
        for line in f:
            stopwords.add(line.strip())
    
    # Ê∑ªÂä†Èü≥‰πêÈ¢ÜÂüüËØçÂÖ∏
    music_terms = ['ÂâØÊ≠å', 'ÁºñÊõ≤', 'ÊóãÂæã', 'ËäÇÂ•è', 'ÂíåÂº¶', 'ÂâçÂ•è', 'Èó¥Â•è', 'Â∞æÂ•è', 
                  'Èü≥Ëâ≤', 'Ê≠åËØç', '‰∏áÊúâÂºïÂäõ', 'Ê±™ËãèÊ≥∑', '‰∏ªÊ≠å', 'bridge', 'hook']
    for term in music_terms:
        jieba.add_word(term)
    
    return stopwords

def chinese_text_processing(text, stopwords):
    words = jieba.cut(text)
    return [word for word in words if word not in stopwords and len(word) > 1]

# 3. TF-IDFÂÖ≥ÈîÆËØçÂàÜÊûê
def tfidf_analysis(df, stopwords, save_dir):
    # ÂáÜÂ§áÊñáÊú¨Êï∞ÊçÆ
    df['ÂàÜËØçÂÜÖÂÆπ'] = df['ËØÑËÆ∫ÂÜÖÂÆπ'].apply(lambda x: chinese_text_processing(x, stopwords))
    df['ÂàÜËØçÊñáÊú¨'] = df['ÂàÜËØçÂÜÖÂÆπ'].apply(' '.join)
    
    # TF-IDFÂàÜÊûê
    tfidf = TfidfVectorizer(max_features=100)
    tfidf_matrix = tfidf.fit_transform(df['ÂàÜËØçÊñáÊú¨'])
    feature_names = tfidf.get_feature_names_out()
    
    # Ëé∑ÂèñÂÖ≥ÈîÆËØçÊùÉÈáç
    word_scores = {word: tfidf_matrix[:, idx].sum() for idx, word in enumerate(feature_names)}
    top_words = sorted(word_scores.items(), key=lambda x: x[1], reverse=True)[:20]
    
    # ‰øùÂ≠òÂÖ≥ÈîÆËØçÁªìÊûú
    keywords_df = pd.DataFrame(top_words, columns=['ÂÖ≥ÈîÆËØç', 'TF-IDFÊùÉÈáç'])
    keywords_df.to_csv(f'{save_dir}/tfidf_keywords.csv', index=False, encoding='utf-8-sig')
    
    # ÂèØËßÜÂåñ
    plt.figure(figsize=(12, 8))
    sns.barplot(x='TF-IDFÊùÉÈáç', y='ÂÖ≥ÈîÆËØç', data=keywords_df)
    plt.title('TF-IDFÊùÉÈáçÊúÄÈ´òÁöÑ20‰∏™ÂÖ≥ÈîÆËØç')
    plt.tight_layout()
    plt.savefig(f'{save_dir}/tfidf_top_keywords.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    # ÁîüÊàêËØç‰∫ë
    all_words = ' '.join(df['ÂàÜËØçÊñáÊú¨'])
    wordcloud = WordCloud(width=800, height=400, 
                         background_color='white',
                         font_path='simhei.ttf',
                         max_words=100).generate(all_words)
    
    plt.figure(figsize=(15, 8))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('ËØÑËÆ∫ÂÖ≥ÈîÆËØçËØç‰∫ë')
    plt.savefig(f'{save_dir}/word_cloud.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    return df, feature_names, tfidf_matrix

# 4. LDA‰∏ªÈ¢òÂª∫Ê®°
def lda_topic_modeling(df, save_dir, num_topics=4):
    # ÂáÜÂ§áLDAËæìÂÖ•Êï∞ÊçÆ
    texts = df['ÂàÜËØçÂÜÖÂÆπ'].tolist()
    dictionary = corpora.Dictionary(texts)
    corpus = [dictionary.doc2bow(text) for text in texts]
    
    # ËÆ≠ÁªÉLDAÊ®°Âûã
    lda_model = models.LdaModel(corpus=corpus,
                               id2word=dictionary,
                               num_topics=num_topics,
                               random_state=42,
                               passes=15,
                               alpha='auto')
    
    # ‰øùÂ≠ò‰∏ªÈ¢òÂÖ≥ÈîÆËØç
    topics = lda_model.print_topics(num_words=8)
    with open(f'{save_dir}/lda_topics.txt', 'w', encoding='utf-8') as f:
        for idx, topic in topics:
            f.write(f"‰∏ªÈ¢ò{idx}: {topic}\n")
    
    # ÂèØËßÜÂåñ‰∏ªÈ¢ò
    vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary)
    pyLDAvis.save_html(vis, f'{save_dir}/lda_visualization.html')
    
    # ‰∏∫ÊØèÊù°ËØÑËÆ∫ÂàÜÈÖç‰∏ªË¶Å‰∏ªÈ¢ò
    df['‰∏ªË¶Å‰∏ªÈ¢ò'] = [sorted(lda_model[doc], key=lambda x: x[1], reverse=True)[0][0] for doc in corpus]
    
    # ‰∏ªÈ¢òÂàÜÂ∏ÉÂèØËßÜÂåñ
    topic_dist = df['‰∏ªË¶Å‰∏ªÈ¢ò'].value_counts().sort_index()
    plt.figure(figsize=(10, 6))
    topic_dist.plot(kind='bar')
    plt.title('ËØÑËÆ∫‰∏ªÈ¢òÂàÜÂ∏É')
    plt.xlabel('‰∏ªÈ¢òÁºñÂè∑')
    plt.ylabel('ËØÑËÆ∫Êï∞Èáè')
    plt.savefig(f'{save_dir}/topic_distribution.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    return df, lda_model

# 5. ÊÉÖÊÑüÂàÜÊûê (Ê®°Êãü)
def sentiment_analysis(df, save_dir):
    # Ê®°ÊãüÊÉÖÊÑüÂàÜÊûê (ÂÆûÈôÖÂ∫îÁî®‰∏≠ÂèØÁî®BosonNLPÊàñÂÖ∂‰ªñAPI)
    np.random.seed(42)
    df['ÊÉÖÊÑüÂàÜÊï∞'] = np.random.normal(loc=0.6, scale=0.2, size=len(df))
    df['ÊÉÖÊÑüÂàÜÊï∞'] = df['ÊÉÖÊÑüÂàÜÊï∞'].clip(0, 1)  # ÈôêÂà∂Âú®0-1ËåÉÂõ¥
    
    # ÊÉÖÊÑüÂàÜÂ∏ÉÂèØËßÜÂåñ
    plt.figure(figsize=(10, 6))
    sns.histplot(df['ÊÉÖÊÑüÂàÜÊï∞'], bins=20, kde=True)
    plt.title('ËØÑËÆ∫ÊÉÖÊÑüÂàÜÊï∞ÂàÜÂ∏É')
    plt.xlabel('ÊÉÖÊÑüÂàÜÊï∞ (0-1)')
    plt.ylabel('È¢ëÊï∞')
    plt.axvline(0.5, color='r', linestyle='--')
    plt.text(0.52, plt.ylim()[1]*0.9, '‰∏≠ÊÄßÈòàÂÄº', color='r')
    plt.savefig(f'{save_dir}/sentiment_distribution.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    # ÊÉÖÊÑü‰∏é‰∏ªÈ¢òÁöÑÂÖ≥Á≥ª
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='‰∏ªË¶Å‰∏ªÈ¢ò', y='ÊÉÖÊÑüÂàÜÊï∞', data=df)
    plt.title('‰∏çÂêå‰∏ªÈ¢òÁöÑÊÉÖÊÑüÂàÜÊï∞ÂàÜÂ∏É')
    plt.xlabel('‰∏ªÈ¢òÁºñÂè∑')
    plt.ylabel('ÊÉÖÊÑüÂàÜÊï∞')
    plt.savefig(f'{save_dir}/sentiment_by_topic.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    # ÊÉÖÊÑü‰∏éÁÇπËµûÊï∞ÁöÑÂÖ≥Á≥ª
    plt.figure(figsize=(12, 6))
    sns.scatterplot(x='ÊÉÖÊÑüÂàÜÊï∞', y='ÁÇπËµûÊï∞', data=df, alpha=0.6)
    plt.title('ÊÉÖÊÑüÂàÜÊï∞‰∏éÁÇπËµûÊï∞ÁöÑÂÖ≥Á≥ª')
    plt.xlabel('ÊÉÖÊÑüÂàÜÊï∞')
    plt.ylabel('ÁÇπËµûÊï∞')
    plt.savefig(f'{save_dir}/sentiment_vs_likes.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    return df

# 6. Êó∂Èó¥Â∫èÂàóÂàÜÊûê
def temporal_analysis(df, save_dir):
    # ÊåâÊúà‰ªΩÁªüËÆ°ËØÑËÆ∫Êï∞Èáè
    monthly_counts = df.groupby('ËØÑËÆ∫Êúà‰ªΩ').size()
    
    plt.figure(figsize=(14, 6))
    monthly_counts.plot()
    plt.title('ËØÑËÆ∫Êï∞ÈáèÈöèÊó∂Èó¥ÂèòÂåñ')
    plt.xlabel('Êó∂Èó¥')
    plt.ylabel('ËØÑËÆ∫Êï∞Èáè')
    plt.grid(True)
    plt.savefig(f'{save_dir}/comments_over_time.png', dpi=300, bbox_inches='tight')
    plt.close()
    
    # ‰∏ªÈ¢òÈöèÊó∂Èó¥ÂèòÂåñ
    monthly_topics = df.groupby(['ËØÑËÆ∫Êúà‰ªΩ', '‰∏ªË¶Å‰∏ªÈ¢ò']).size().unstack()
    monthly_topics.plot(kind='area', stacked=True, figsize=(14, 6))
    plt.title('‰∏ªÈ¢òÈöèÊó∂Èó¥ÂèòÂåñ')
    plt.xlabel('Êó∂Èó¥')
    plt.ylabel('ËØÑËÆ∫Êï∞Èáè')
    plt.legend(title='‰∏ªÈ¢ò')
    plt.savefig(f'{save_dir}/topics_over_time.png', dpi=300, bbox_inches='tight')
    plt.close()

# ‰∏ªÂáΩÊï∞
def main():
    # Âä†ËΩΩÊï∞ÊçÆ
    filepath = 'netease_comments_165339.csv'
    df = load_and_preprocess_data(filepath)
    
    # ËÆæÁΩÆ‰∏≠ÊñáÂàÜÊûêÂô®
    stopwords = setup_chinese_analyzer()
    
    # ÊâßË°åÂàÜÊûêÊµÅÁ®ã
    df, feature_names, tfidf_matrix = tfidf_analysis(df, stopwords, result_dir)
    df, lda_model = lda_topic_modeling(df, result_dir)
    df = sentiment_analysis(df, result_dir)
    temporal_analysis(df, result_dir)
    
    # ‰øùÂ≠òÂ§ÑÁêÜÂêéÁöÑÊï∞ÊçÆ
    df.to_csv(f'{result_dir}/processed_comments.csv', index=False, encoding='utf-8-sig')
    print(f"ÂàÜÊûêÂÆåÊàêÔºåÁªìÊûúÂ∑≤‰øùÂ≠òÂà∞: {result_dir}")

if __name__ == "__main__":
    main()
```
